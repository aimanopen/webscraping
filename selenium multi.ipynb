```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv

# Set up the Chrome webdriver
service = Service(ChromeDriverManager().install())

# Set up Chrome options with the executable path
chrome_options = webdriver.ChromeOptions()
# Example user-agent change
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")

# Create an instance of the Chrome webdriver with options
driver = webdriver.Chrome(service=service, options=chrome_options)

# Loop through the pages
for page_number in range(1, 21):  # Assuming there are 20 pages
# Navigate to the current page
    driver.get(f'https://www.amazon.com/s?k=laptop&crid=2E33KGYSKVIYE&sprefix=laptop%2Caps%2C128&ref=nb_sb_noss_1&page={page_number}')
# Wait for elements to load (simple sleep, WebDriverWait is better in practice)
time.sleep(5)
```

```python
#Initialize the list to store all data
all_laptop_data = []

# Get the page content
content = driver.page_source

# Parse the page with BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

# Extract laptop information (modify this part based on the actual structure of the page)
products = soup.find_all('h2', {'class': 'a-size-mini'})
prices = soup.find_all('span', {'class': 'a-price-whole'})

# Initialize the list to store data for the current page
laptop_data = []

for product, price in zip(products, prices):
    laptop_data.append({'Product': product.text, 'Price': price.text})

# Add the data for the current page to the overall list
all_laptop_data.extend(laptop_data)

# Save data to CSV file
csv_file_path = 'laptop_data_all_pages.csv'
headers = ['Product', 'Price']

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=headers)
    writer.writeheader()
    writer.writerows(all_laptop_data)

# Close the WebDriver when done
driver.quit()
```

```python
import pandas as pd
df=pd.read_csv('laptop_data_all_pages.csv')
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
df['Product']=df['Product'].str.slice(0,20)
df.head()
```

```python
df = df.dropna()
pd.set_option('display.max_rows', None)
df
```

```python
df = df[~df['Product'].astype(str).str.contains('-')]
#reset ID column
df.reset_index(drop=True, inplace=True)
pd.set_option('display.max_rows', None)
df
```

```python
#Remove text from the Price column: 
df['Price'] = df['Price'].replace('[^\d.]', '', regex=True)
pd.set_option('display.max_rows', None)
df
```

```python
#Convert 'Price' column to numeric coercing errors (this turns non-convertible values into NaN)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'] = df['Price'].astype(float)
#Use price over $100:
df = df[df['Price'] >= 100]
#Use price below $2000
df = df[df['Price'] < 2000]
#We can make a single line of code: df=df[(df['Price'] >=100)&(df['Price'] <2000)]
pd.set_option('display.max_rows', None)
df
```

```python
df.reset_index(drop=True, inplace=True)
pd.set_option('display.max_rows', None)
df
```
